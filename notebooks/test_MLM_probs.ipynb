```python
import sys
sys.path.append("..")
import json
from streaming.base.format import reader_from_json
import os
import numpy as np
from streaming.base.spanner import Spanner
import h5py
from filelock import FileLock

local = "/mnt/nfs_dna/minja/DNALM/promoter_pretrain/small_test_100_human_promoters/mds/"
split = "train"

index_file_path = os.path.join(local, split, "index.json")
obj = json.load(open(index_file_path))
shards = []
for info in obj["shards"]:
	shard = reader_from_json(local, split, info)
	raw_filename = os.path.join(shard.dirname, shard.split, shard.raw_data.basename)
	assert os.path.isfile(raw_filename), f"Raw file {raw_filename} does not exist"
	shard.validate(True)
	shards.append(shard)
samples_per_shard = np.array([shard.samples for shard in shards], np.int64)
l = samples_per_shard.sum()
spanner = Spanner(samples_per_shard)


h5_dirs = "../runs/mlm_efficiency/test_with_mlm_2/train"
nanmeans = []
with open("output_fwd.bedGraph", "w") as bg_fwd, open("output_rev.bedGraph", "w") as bg_rev, open("output.bed", "w") as bed:
	for index in range(l):
		shard_id, shard_sample_id = spanner[index]
		shard = shards[shard_id]
		sample = shard[shard_sample_id]
		assert sample["file_id"] == "GCF_000001405"
		if not os.path.exists(os.path.join(h5_dirs, f"shard_{shard_id}.hdf5")):
			continue
		hdf5_file = os.path.join(h5_dirs, f"shard_{shard_id}.hdf5")
		with FileLock(hdf5_file+".lock"):
			with h5py.File(hdf5_file, "r") as f:
				if not str(shard_sample_id) in f:
					continue
				promoters_dir = f"/mnt/nfs_dna/minja/DNALM/promoter_pretrain/small_test_100_human_promoters/jsons/GCF_000001405.40"
				promoters_file = os.path.join(promoters_dir, f"GCF_000001405.40.jsonl")

				assert os.path.exists(promoters_file), f"Promoter file {promoters_file} not found"
				promoter_file_lines = open(promoters_file, 'r').readlines()
				line = promoter_file_lines[sample['line_id']]
				line = json.loads(line)
				# print (line.keys())
				probabilities = np.array(f[str(shard_sample_id)][:])
				offset = sample['chunk_offset']
				# print (len(probabilities), offset)
				chrom = line['chromosome']
				if chrom not in chroms[["RefSeq_seq_accession"]].values:
					continue
				chromlen = int(chroms[chroms["RefSeq_seq_accession"] == chrom]["Seq_length"].values[0])
				chrom_renamed = chroms[chroms["RefSeq_seq_accession"] == chrom]["UCSC_style_name"].values[0]

				strand = line['sample_strand']
				if strand == "-":
					# probabilities = probabilities[::-1]
					start = line['start']+len(line['text'][0]) - offset - 1
					strand = -1
					bg = bg_rev
				elif strand == "+":
					start = line['start'] + offset
					strand = 1
					bg = bg_fwd
				else:
					raise ValueError(f"Invalid strand: {strand}")

				assert start <= chromlen, f"Start {start} is greater than chromosome length {chromlen}: {chrom} {chrom_renamed} {start} {offset} {len(line['text'][0])} {strand}"
				promoter_id = line['gene_ID']+':'+line['transcript_ID']
				sample_strand = line['sample_strand']
				# print (chrom, start, end, promoter_id, sample_strand)
				probabilities = 1 - probabilities
				nanmeans.append(np.nanmean(probabilities))
				probabilities = np.nan_to_num(probabilities, nan=-1)
				for i in range(probabilities.shape[0]):
					bg.write(f"{chrom_renamed}\t{start+i*strand}\t{start+i*strand+1}\t{probabilities[i]}\n")
				if strand == 1:
					bed.write(f"{chrom_renamed}\t{start}\t{start+len(probabilities)}\t{promoter_id}\t{sample_strand}\n")
				else:
					bed.write(f"{chrom_renamed}\t{start-len(probabilities)}\t{start}\t{promoter_id}\t{sample_strand}\n")
print (np.nanmean(nanmeans))
```
